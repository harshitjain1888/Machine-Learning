```python
class employees:
  def __init__(self,emp_id,name,salary):
    self.name=name
    self.salary=salary
    self.emp_id=emp_id

employees= {
    101:employees(101,'john',5000),
    102:employees(102,'jane',6000),
    103:employees(103,'jack',7000)
}

for emp_id, emp in employees.items():
  print(f"Employee ID: {emp.emp_id}")
  print(f"Name: {emp.name}")
  print(f"Salary: {emp.salary}")
```
